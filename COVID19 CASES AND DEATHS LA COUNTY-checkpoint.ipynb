```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the full file path to your CSV file
desktop_path = r'C:\Users\celes\Desktop'
file_name = 'LA_County_COVID_Cases_20231018.csv'
file_path = f'{desktop_path}\\{file_name}'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Convert the 'date' column to a datetime data type
df['date'] = pd.to_datetime(df['date'])

# Sort the DataFrame by date
df = df.sort_values(by='date')

# Select the first 10 rows
df_first_10 = df.head(10)

# Create a line plot for COVID-19 cases and deaths for the first 10 rows
plt.figure(figsize=(12, 6))
plt.plot(df_first_10['date'], df_first_10['cases'], label='Total Cases', color='blue')
plt.plot(df_first_10['date'], df_first_10['deaths'], label='Total Deaths', color='red')

# Customize the plot
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('COVID-19 Cases and Deaths Over Time (First 10 Rows)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

#This plot showed the overall trends in COVID-19 cases and deaths over time in Los Angeles County. It visualized how the counts of cases and deaths changed from an earlier date to a later date.

#There is a connection between an increase in COVID-19 cases and a subsequent rise in deaths. However, it also suggested that there might be a time delay between the increase in cases and deaths.
